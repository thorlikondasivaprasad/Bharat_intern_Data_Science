```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection && preprocessing

```python
titanic_data = pd.read_csv("train.csv")
display(titanic_data.head())
```

```python
# getting the shape of the data
titanic_data.shape
```

```python
# getting some info about the data

titanic_data.info()
```

```python
# checking the no.of null values in each column

titanic_data.isnull().sum()
```

Handling Missing values

```python
# here drop the cabin column from the dataframe

titanic_data = titanic_data.drop(columns="Cabin" , axis=1)
```

```python
titanic_data.head() # here cabin col is removed
```

```python
# filling the missing values in the "Age" column with the mean value -- data.fillna()

titanic_data["Age"].fillna(titanic_data["Age"].mean(),inplace=True)
```

```python
# filling missing values in "Embarked" ,which  is string data so,it'll be replacer with the "mode" value

print(titanic_data["Embarked"].mode()[0])  # here "S" is repeated more no.of times
```

```python
titanic_data["Embarked"].fillna(titanic_data["Embarked"].mode()[0] , inplace=True)
```

```python
# check for null values

titanic_data.isnull().sum()

# here there is no missing values
```

Label Encodding - for Categorical columns "Sex" ,"Embarked".

```python
titanic_data.head()
```

```python
titanic_data['Sex'].value_counts()
```

```python
titanic_data['Embarked'].value_counts()
```

```python
# converting Categorical columns

titanic_data.replace({"Sex":{"male":0,"female":1},
                     "Embarked":{"S":0,"C":1,"Q":2}},
                     inplace=True)
```

```python
titanic_data.head(6)
```

Separating  ,
 Features-->Sex	,Age,....                               
 Target --> Survived

```python
X = titanic_data.drop(columns=["PassengerId","Name","Ticket","Survived"],axis=1)
Y = titanic_data["Survived"]
```

```python
display(X)
```

```python
display(Y)
```

Splitting the data into Training data and Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)  # 80% ->training data, 20% ->testing data

```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Building                               
LogisticRegression

```python
model = LogisticRegression()
```

```python
# training the model

model.fit(X_train ,Y_train)
```

Evaluate the model                        
Accuracy_score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
```

```python
print(X_train_prediction,"\n")
print(X_train_prediction.shape)
```

```python
training_data_accuracy =accuracy_score(Y_train,X_train_prediction)
print("Accuracy Score of Training Data : ",training_data_accuracy)
```

Evaluate the model with the Test data

```python
# Evaluate the model with the Test data

X_test_prediction = model.predict(X_test)
```

```python
print(X_test_prediction,"\n")
print(X_test_prediction.shape)
```

```python
test_data_accuracy = accuracy_score(Y_test,X_test_prediction)
print("Accuracy Score of Testing Data : ",test_data_accuracy)
```

```python
# here both acc.scores are similar to each other , and not prone to overfitting or underfitting
```
